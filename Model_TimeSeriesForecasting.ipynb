```python
# Seasonal differencing

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
from statsmodels.graphics.tsaplots import plot_acf

# Load and preprocess the data
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Select the columns of interest
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close']

# Iterate over each column
for column in columns:
    # Select the column as the target variable
    price_series = data[column]

    # Perform seasonal differencing with lag 12
    seasonal_diff = price_series.diff(12)
    seasonal_diff.dropna(inplace=True)

    # Plot the autocorrelation function (ACF) of the seasonal differenced series
    plot_acf(seasonal_diff, lags=30)
    plt.title(f'Autocorrelation Function ({column})')
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()

    # Split the data into train and test sets
    train_size = int(len(price_series) * 0.8)
    train_data, test_data = price_series[:train_size], price_series[train_size:]

    # Define the SARIMA model
    model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))

    # Fit the model to the training data
    model_fit = model.fit()

    # Perform the time series forecasting on the test data
    forecast = model_fit.forecast(steps=len(test_data))

    # Print the column name
    print("Column:", column)

    # Print the forecasted values
    print("Forecasted values:")
    print(forecast)

    # Evaluate the model using mean squared error (MSE)
    mse = mean_squared_error(test_data, forecast)
    print("Mean Squared Error (MSE):", mse)

    # Print a separator
    print("---------------------------")
```
